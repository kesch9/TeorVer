### Задание 1
Из колоды в 52 карты извлекаются случайным образом 4 карты. 
a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.
2) На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?
3) В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?
4) В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

```python
import math
C13 = math.factorial(13)/(math.factorial(4)*math.factorial(13-4))
C52 = math.factorial(52)/(math.factorial(4)*math.factorial(52-4))
P = C13/C52
print("Вероятность того, что все карты – крести = ", P)
C48 = math.factorial(48)/(math.factorial(4)*math.factorial(48-4))
C1234 = C52-C48
P = C1234/C52
print("Bероятность, что среди 4-х карт окажется хотя бы один туз = ", P)
```

### Задание 2
На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

```python
C10 = math.factorial(10)/(math.factorial(3)*math.factorial(10-3))
P = 1/C10
print("Вероятность того, что человек, не знающий код, откроет дверь с первой попытки = ", P)
```

### Задание 3
В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

```python
C9 = math.factorial(9)/(math.factorial(3)*math.factorial(9-3))
C15 = math.factorial(15)/(math.factorial(3)*math.factorial(15-3))
P = C9/C15
print("Вероятность того, что все извлеченные детали окрашены = ", P)
```

### Задание 4
 В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

```python
C100 = math.factorial(100)/(math.factorial(2)*math.factorial(100-2))
P = 1/C100
print("Вероятность того, что 2 приобретенных билета окажутся выигрышными = ", P)
```
